```
import gym
from stable_baselines.common.vec_env import DummyVecEnv
from stable_baselines.deepq.policies import MlpPolicy, CnnPolicy
from stable_baselines import DQN
#get cartpole from gym
env = gym.make('CartPole-v0')
env = DummyVecEnv([lambda: env])
# Policy object that implements DQN policy, using a MLP (2 layers of 64)
model = DQN(MlpPolicy, env, verbose=1)
model.learn(total_timesteps=100000)
#   The process gets started by calling reset(), which returns an initial observation.
obs = env.reset()
i = 1
#     This will run an instance of the CartPole-v0 environment for 1000 timesteps at each step.
while (i < 1000):
    action, _states = model.predict(obs)
    obs, rewards, dones, info = env.step(action)
    print(rewards," ",i);
    i =i+ 1
```
