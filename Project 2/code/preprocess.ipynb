```python
#from sklearn.cluster import KMeans
import numpy as np
import csv
import math
import matplotlib.pyplot
from matplotlib import pyplot as plt
import pandas as pd

##generate key value pair of data set; key:imgae_id,value:features
def GenerateMapForGSC(filePath):
    Map = {}
    with open(filePath, 'rU') as fi:
        reader = csv.reader(fi);
        next(reader);
        for row in reader:
            i=0;
            list=[];
            for column in row:
                if(i==0):
                    key=column;
                elif(i>0) :
                    list.append(int(column));
                i=i+1
            Map[key] = list;
            #print(Map[key]);
    return Map;

##returns subtraction and concatination list of given csv image_id pairs
def GenerateConcatList(filePath,KVMap):
    Map = {}
    with open(filePath, 'rU') as fi:
        reader = csv.reader(fi);
        next(reader);
        sub = [];
        target = [];
        conca = [];
        for row in reader:
            i=0;
            c1=[];
            c2=[];
            for column in row:
                if(i==2):
                    target.append(int(column));
                elif(i==0) :
                    c1=KVMap[column];
                elif(i==1):
                    c2 = KVMap[column];
                i = i + 1;
            myList=c1+c2;
            myList.extend([int(column)])
            conca.append(myList);
            subtract=np.subtract(c1,c2);
            subtract=np.append(subtract,int(column));
            sub.append(subtract);

        Map["target"] = target;
        Map["sub"] = sub;
        Map["concat"] = conca;
            #print(Map[key]);
    return Map;


KVData = GenerateMapForGSC('/Users/vishal/PycharmProjects/ML2/GSC-Features.csv');
#print(KVData)
#print(len(KVData))
SamePairKV = GenerateConcatList('/Users/vishal/PycharmProjects/ML2/same_pairs_gsc.csv',KVData);
#print(len(SamePairKV["concat"]))
DiffPairKV = GenerateConcatList('/Users/vishal/PycharmProjects/ML2/diffn_pairs_gsc.csv',KVData);
#print(len(DiffPairKV["concat"]))
#taking 791 rows of same pair and 791 rows of different pair,balancing data
sub=SamePairKV["sub"] + DiffPairKV["sub"][:71531];
#taking 791 rows of same pair and 791 rows of different pair,balancing data
concat=SamePairKV["concat"] + DiffPairKV["concat"][:71531];

target=SamePairKV["target"] + DiffPairKV["target"][:71531];
print(target);
print(sub);
print(concat);
#shuffle list for randomising rows
np.random.shuffle(sub);
np.random.shuffle(target);
np.random.shuffle(concat);

df1 = pd.DataFrame(concat);
df2 = pd.DataFrame(sub);
# print(df2)

targetConcat = []
targetConcat = df1[df1.columns[-1]]

targetSub = []
targetSub = df2[df2.columns[-1]]

targetConcat = pd.DataFrame(targetConcat)
targetSub = pd.DataFrame(targetSub)

#concat
df1 = df1.drop(df1.columns[1024], axis=1)
#subtract
df2 = df2.drop(df2.columns[512], axis=1)

targetConcat.to_csv('GSCtargetConcat.csv',header=None,index=False)

targetSub.to_csv('GSCtargetSub.csv',header=None,index=False)

df1.to_csv('GSCfeaturesConcat.csv',header=None,index=False)

df2.to_csv('GSCfeaturesSub.csv',header=None,index=False)
```
